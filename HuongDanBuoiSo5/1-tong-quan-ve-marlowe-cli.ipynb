## Tổng quan về Marlowe CLI

### Chức năng của marlowe-cli

`marlowe-cli` là một công cụ gồm các các câu lệnh (command-line interface) được dùng để triển khai và thực hiện hợp đồng Marlowe trên mạng testnet và mainnet

_Chúng ta sử dụng các chương trình như Command Prompt, PowerShell trong hệ điều hành Microsoft Windows hoặc Terminal trong hệ điều hành Linux và macOS để chạy các câu lệnh có trong công cụ `marlowe-cli`_

Tương tự như công cụ `cardano-cli` cho phép thực hiện các giao dịch đơn giản và thực hiện các Plutus scripts trên blockchain, công cụ `marlowe-cli` hỗ trợ chúng ta tạo ra các giao dịch để triển khai và tương tác với hợp đồng Marlowe. 

_Ví dụ: chúng ta có thể dùng `marlowe-cli` để tạo ra giao dịch, đo kích thước của giao dịch, gửi giao dịch lên mạng lưới, thử nghiệm quá trình thực hiện hợp đồng (off-chain), hoặc tích hợp công cụ `marlowe-cli` vào các chương trình / phần mềm / thư viện khác._

### Mô hình eUTxO

### Các Plutus Scripts thường được sử dụng để thực hiện một hợp đồng Marlowe

| Plutus Validator Script | Tác dụng | Datum | Redeemer |
|-----------|-------|----------|----------|
| Role Token Minting Script | Tạo ra các role tokens | n/a | n/a |
| Marlowe Application Script | Kiểm tra hành động của các bên tham gia có hợp lệ hay không | Các phần chưa được thực hiện của hợp đồng; Số dư của các tài khoản nội bộ; Lịch sử những hành động đã được thực biển bởi các bên tham gia; Giá trị các tham số | Hành động tiếp theo của các bên tham gia (_ví dụ: deposit, choice_) |
| Payout Script|  Trả tiền cho các bên tham gia | Vai trò được trả tiền | n/a |

### Role token

- Mỗi hợp đồng Marlowe sẽ có một currency symbol (policy id) riêng cho các role tokens của nó. Đồng thời, mỗi hợp đồng sẽ có một Plutus validator script "Payout" để nhận tiền từ hợp đồng và trả cho các bên tham gia.
- Các bên tham gia thường tương tác với hợp đồng Marlowe bằng cách tạo các giao dịch và sử dụng role token để xác thực vai trò cũng như quyền hạn của họ.
- Các bên tham gia sẽ rút tiền từ một Plutus validator script đặc biệt "Payout" bằng cách sử dụng role token của họ.
    - Datum của script này chỉ đơn giản là tên của vai trò (role) được phép rút số tiền có trong UTxO đó

### Vòng đời của một hợp đồng Marlowe

### Các câu lệnh của marlowe-cli

Các câu lệnh trong công cụ `marlowe-cli` được chia làm 2 nhóm chính, gồm nhóm lệnh cấp cao và nhóm lệnh cấp thấp. 
- Các câu lệnh cấp cao cho phép người dùng tạo ra hợp đồng Marlowe từ các mẫu hợp đồng có sẵn hoặc sử dụng hợp đồng chúng ta tạo ra trên Marlowe Playground. Sau đó người dùng sẽ cung cấp các thông tin đầu vào tại từng bước thực thi hợp đồng. Việc chuyển đổi trạng thái của hợp đồng cũng như tạo và gửi các giao dịch sẽ được thực hiện tự động bởi công cụ. Người dùng không cần phải quan tâm đến những thứ liên quan đến Plutus như datum, redeemer, validator.
- Các câu lệnh cấp thấp cho phép người dùng can thiệp sâu hơn vào quá trình tạo ra các giao dịch để thực hiện hợp đồng. Người dùng có thể sử dụng các câu lệnh cấp thấp để tạo ra các files chứa thông tin về Plutus validator, datum, redeemer và contract address, sau đó kết hợp các files này với nhau để tạo thành một giao dịch và gửi giao dịch đó lên mạng lưới.

```bash
marlowe-cli --help
```
